These samples demonstrate the structure of the nodes and edges that were prepared for ingestion into Neo4j. Each node has an `id`, `type`, and potentially other properties like `text` or `name`, along with the `properties` dictionary for `MealEvent` nodes. Each edge specifies a `source` node, a `target` node, and the `type` of relationship.

```python
import pandas as pd
import time
from datasets import load_dataset
from gliner import GLiNER
from tqdm import tqdm
from neo4j import GraphDatabase


URI = "neo4j+s://faa5d1f5.databases.neo4j.io"
AUTH = ("neo4j", "cPjEt08Kq32nYEC2YX80W1YoknpST6gTBnKmNXw_VEs")

print("Loading Dataset and Model...")
ds = load_dataset("dongx1997/NutriBench", "v2")
df = ds['train'].to_pandas().head(100)

model = GLiNER.from_pretrained("urchade/gliner_medium-v2.1")
labels = ["Food", "MealTime", "Unit"]

all_nodes = {}
all_edges = []

print("Extracting Entities...")
for index, row in tqdm(df.iterrows(), total=len(df)):
    text_content = str(row.get('meal_description', ''))
    if len(text_content) < 3: continue

    meal_id = f"meal_{index}"
    nutrients = {
        "calories": row.get('energy', 0),
        "protein": row.get('protein', 0),
        "fat": row.get('fat', 0),
        "carb": row.get('carb', 0)
    }

    if meal_id not in all_nodes:
        all_nodes[meal_id] = {
            "id": meal_id,
            "type": "MealEvent",
            "text": text_content,
            "properties": nutrients # داده‌های تغذیه‌ای
        }

    entities = model.predict_entities(text_content, labels, threshold=0.3)
    seen_entities_in_this_row = set()

    for entity in entities:
        e_text = entity['text'].lower().strip()
        e_label = entity['label']
        node_id = f"{e_label}_{e_text}".replace(" ", "_")

        if node_id in seen_entities_in_this_row: continue
        seen_entities_in_this_row.add(node_id)

        if node_id not in all_nodes:
            all_nodes[node_id] = {
                "id": node_id, "type": e_label, "name": e_text
            }

        edge_type = "CONTAINS" if e_label == "Food" else "RELATED_TO"
        all_edges.append({"source": meal_id, "target": node_id, "type": edge_type})

print(f"Extraction Done. Nodes: {len(all_nodes)}, Edges: {len(all_edges)}")

```

```python
# ==========================================
driver = GraphDatabase.driver(URI, auth=AUTH)

def create_nodes_batch(tx, batch_nodes):
    for node_data in batch_nodes:
        label = node_data['type']
        params = node_data.copy()

        if 'properties' in params:
            nutrients = params.pop('properties')
            params.update(nutrients)
        if 'text' not in params:
            params['text'] = params.get('name', '')

        query = f"MERGE (n:{label} {{id: $id}}) SET n += $params"
        tx.run(query, params=params, id=node_data['id'])

def create_edges_batch(tx, batch_edges):
    for edge in batch_edges:
        query = f"""
        MATCH (a {{id: $source}})
        MATCH (b {{id: $target}})
        MERGE (a)-[r:{edge['type']}]->(b)
        """
        tx.run(query, source=edge['source'], target=edge['target'])

with driver.session() as session:
    session.run("MATCH (n) DETACH DELETE n")

BATCH_SIZE = 200
node_list = list(all_nodes.values())
print(f"Ingesting {len(node_list)} nodes to Neo4j...")

for i in range(0, len(node_list), BATCH_SIZE):
    batch = node_list[i : i + BATCH_SIZE]
    try:
        with driver.session() as session:
            session.execute_write(create_nodes_batch, batch)
    except Exception as e:
        print(f"Error in node batch {i}: {e}")
        time.sleep(2)

print(f"Ingesting {len(all_edges)} edges to Neo4j...")
for i in range(0, len(all_edges), BATCH_SIZE):
    batch = all_edges[i : i + BATCH_SIZE]
    try:
        with driver.session() as session:
            session.execute_write(create_edges_batch, batch)
    except Exception as e:
        print(f"Error in edge batch {i}: {e}")
        time.sleep(2)

driver.close()
print(" All Data Successfully Ingested into Neo4j!")
```

```python
from langchain_community.vectorstores import Neo4jVector
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

URI = "neo4j+s://faa5d1f5.databases.neo4j.io"
AUTH = ("neo4j", "cPjEt08Kq32nYEC2YX80W1YoknpST6gTBnKmNXw_VEs")
OPENROUTER_API_KEY = "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"

print("Creating Vector Index...")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vector_store = Neo4jVector.from_existing_graph(
    embedding=embeddings,
    url=URI,
    username=AUTH[0],
    password=AUTH[1],
    index_name="meal_description_vector",
    node_label="MealEvent",
    text_node_properties=["text"],
    embedding_node_property="embedding",
)

llm = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
    model="google/gemma-3-12b-it:free",
    temperature=0,
)

template = """
[Instructions]
به عنوان متخصص تغذیه، فقط بر اساس داده‌های زیر پاسخ بده:
{context}

سوال کاربر: {question}
"""
prompt = ChatPromptTemplate.from_template(template)
retriever = vector_store.as_retriever(search_kwargs={'k': 3})

rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

print("\nTesting Query:")
print(rag_chain.invoke(" دستور \ختشکدام غذا بیشترین پروتئین را داشت؟دستور پختش. اسمش چیه"))
```
